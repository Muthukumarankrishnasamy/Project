```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#2.1- Analyse the data using pandas
amaz = pd.read_csv(r"c:\Users\dhanz\OneDrive\Desktop\Assignment\electronics.csv", header = 0)
amaz

#2.2- Convert the columns brand,rating,cateogory,user_id,item_id as string
amaz[['brand', 'rating', 'category', 'user_id', 'item_id']] = amaz[['brand', 'rating', 'category', 'user_id', 'item_id']].astype(str)

#2.3- Drop all duplicate values(dropna) in rating cateogory
df = amaz.drop_duplicates().dropna()

#2.4- check for missing values(isnull)
missing_values = amaz.isnull().sum()

#2.5- Find the distribution of ratings using countplot,Histogram.
plt.figure(figsize=(8, 6))
sns.countplot(amaz, x='rating')
plt.title('Rating Distribution (Countplot)')
plt.show()
plt.figure(figsize=(8, 6))
plt.hist(amaz['rating'], bins=10)
plt.title('Rating Distribution (Histogram)')
plt.xlabel('Rating')
plt.ylabel('frequency')
plt.show()

```

```python
#2.6-Find the best year of the rating
#2.7-Create a bar chart of for the best year of the rating
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
amaz = pd.read_csv(r"c:\Users\dhanz\OneDrive\Desktop\Assignment\electronics.csv", header = 0)
amaz
year_rating = amaz.groupby('year')['rating'].sum()
best_year = year_rating.idxmax()
print(f"The best year of rating is: {best_year}")
plt.figure(figsize=(8, 6))
plt.bar(year_rating.index, year_rating.values, color='skyblue')
plt.title('Rating by Year')
plt.xlabel('Year')
plt.ylabel('Total Rating')
plt.xticks(year_rating.index,rotation=75, ha='right', size=8)
plt.show()

#2.8- What brand sold the most in 2015 and do a bar chart
amaz_2015 = amaz[amaz['year'] == 2015]
brand_2015 = amaz_2015.groupby('brand')['rating'].sum()
selling_brand = brand_2015.idxmax()
plt.figure(figsize=(8, 6))
plt.bar(brand_2015.index, brand_2015.values, color='lightgreen')
plt.title('Brand sold in 2015')
plt.xlabel('brand')
plt.ylabel('Rating_sum')
plt.xticks(rotation=75, ha='right', size=8)
plt.show()
print(f"The brand that sold the most in 2015 is: {selling_brand}")

#2.9- what product sold the most in 2016 and do a barchart
amaz_2016 = amaz[amaz['year'] == 2016]
product_2016 = amaz_2016.groupby('category')['rating'].sum()
selling_product = product_2016.idxmax()
plt.figure(figsize=(8, 6))
plt.bar(product_2016.index, product_2016.values, color='lightgreen')
plt.title('Product sold in 2016')
plt.xlabel('Category')
plt.ylabel('Rating_sum')
plt.xticks(rotation=45, ha='right')
plt.show()
print(f"The product that sold the most in 2016 is: {selling_product}")

#2.10-what product sold the most in 2017 and do a barchart
amaz_2017 = amaz[amaz['year'] == 2017]
product_2017 = amaz_2017.groupby('category')['rating'].sum()
selling_product = product_2017.idxmax()
plt.figure(figsize=(8, 6))
plt.bar(product_2017.index, product_2017.values, color='lightblue')
plt.title('Product sold in 2017')
plt.xlabel('Category')
plt.ylabel('Rating_sum')
plt.xticks(rotation=45, ha='right')
plt.show()
print(f"The product that sold the most in 2017 is: {selling_product}")


#2.11-category percentage sales and do a pie chart
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
amaz = pd.read_csv(r"c:\Users\dhanz\OneDrive\Desktop\Assignment\electronics.csv", header = 0)
amaz
category_sum = amaz.groupby('category')['rating'].sum()
category_percentage= (category_sum / category_sum.sum()) * 100

# Create a pie chart for category percentage sales
plt.figure(figsize=(8, 8))
plt.pie(category_percentage, labels=None, autopct='%1.1f%%', startangle=90, colors=['orange', 'lightgreen', 'lightblue','red','green','yellow'])
plt.legend(category_percentage.index, title='Categories', loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Category Percentage')
plt.show()

```
