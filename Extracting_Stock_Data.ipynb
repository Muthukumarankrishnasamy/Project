```python
import yfinance as yf
import pandas as pd
import json
import matplotlib.pyplot as plt

# Define the stock symbol
stock_symbol = "AAPL"

# Create a Ticker object for the specified stock symbol
stock = yf.Ticker(stock_symbol)

# Extract stock information
stock_info = stock.info
print("Stock Information:")
print(json.dumps(stock_info, indent=2))

# Extract and print the country of the stock
country = stock_info.get('country', 'N/A')
print("\nCountry:", country)

# Extract historical share price data
share_data = stock.history(period="max")

# Reset the index of the DataFrame
share_data.reset_index(inplace=True)

# Display and plot the historical share price data
print("\nHistorical Share Price Data:")
print(share_data.head())
share_data.plot(x="Date", y="Open", title=" Historical Share Price")
plt.show()

# Extract and plot historical dividends data
dividend_data = stock.dividends
print("\nDividends Data:")
print(dividend_data.head())
dividend_data.plot(title=f"{stock_symbol} Dividends Over Time")
plt.show()
```
